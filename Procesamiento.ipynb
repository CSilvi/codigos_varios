```python
import pandas as pd
import sqlite3
import matplotlib
```

```python
# Load to csv
csv = 'sysarmy_survey_2020_processed.csv'
df = pd.read_csv(csv)
df[:5]
```

```python
df= df.iloc[:10]
df
```

```python
# Creating/Connect SQLite database
connection = sqlite3.connect('database.db')
cursor = connection.cursor()

# Insert the data from the DataFrame into the SQLite table
df.to_sql('new_db_sql',connection,if_exists='replace')

# Printing pandas dataframe
pd.read_sql('''SELECT * FROM new_db_sql''', connection)
```

```python
res = cursor.execute("SELECT name FROM sqlite_master")
res.fetchone()
```

## Funciones 

```python
cursor.execute("PRAGMA table_info(new_db_sql)")
# Fetch all column names and store them in a list
columns = cursor.fetchall()
# Print each column name
for column in columns:
    print(column[1])
```

```python
# Renames a table column
def rename_column (table_name, old_column_name, new_column_name):
    cursor.execute(f"ALTER TABLE {table_name} RENAME COLUMN {old_column_name} TO {new_column_name}")
    return pd.read_sql(f"SELECT * FROM {table_name}", connection)
```

```python
rename_column('new_db_sql', 'profile_gender', 'profile_gender_')
```

```python
# Add Columm a table
def add_columns (table_name, column_name):
    cursor.execute(f"ALTER TABLE {table_name} ADD COLUMN {column_name}")
    return pd.read_sql(f"SELECT * FROM {table_name}", connection)
   
```

```python
add_columns ('new_db_sql', 'new_column')
```

```python
cursor.execute("PRAGMA table_info(new_db_sql)")
# Fetch all column names and store them in a list
columns = cursor.fetchall()
# Print each column name
for column in columns:
    print(column[1])
```

```python
#  drop columns a table
def drop_columns (table_name, column_name):
    cursor.execute(f"ALTER TABLE {table_name} DROP COLUMN {column_name}")
    return pd.read_sql(f"SELECT * FROM {table_name}", connection)
```

```python
drop_columns('new_db_sql', 'new_column')
```

```python
cursor.execute("PRAGMA table_info(new_db_sql)")
# Fetch all column names and store them in a list
columns = cursor.fetchall()
# Print each column name
for column in columns:
    print(column[1])
```
